```python
import csv

with open('insurance.csv', 'r') as insurance_csv:
    insurance_reader = csv.reader(insurance_csv)
    header = next(insurance_reader)
    rows = []
    for row in insurance_reader:
        rows.append(row)  
```

```python
insurance_dict = {}
count = 1
for row in rows:
    dict_row = { "Age": row[0], "Gender": row[1], "BMI": row[2], "Children": row[3], "Smoker": row[4], "Region": row[5], "Cost": row[6] }
    insurance_dict[count] = dict_row
    count += 1
```

```python
smokers = []
for row in rows:
    smokers.append(row[4])
age = []
for row in rows:
    age.append(row[0])

bmi = []
for row in rows:
    bmi.append(row[2]) 

total_people = len(smokers)

smokers_age = list(zip(age, smokers))
smokers_bmi = list(zip(bmi, smokers))
age_bmi = list(zip(age, bmi))
combined_columns = list(zip(age, smokers, bmi))
```

```python
# Determining the percentage of smokers in the insurance info data 
```

```python
def smoker_percentage(smokers):
    count = 0
    for smoker in smokers:
        if smoker == 'yes':
            count += 1
    percentage = round(count / total_people * 100, 2)
    return print("There are " + str(percentage) + "% smokers in the dataset.")
```

```python
# Determining the average age of smokers/non smokers
```

```python
def smoker_age(list):
    total_age = 0
    smokers = 0
    for entry in list:
        if entry[1] == 'yes':
            curr_age = int(entry[0])
            total_age += curr_age
            smokers += 1
    avg_age_of_smoker = round(total_age / smokers, 2)
    return print("The average age of a smoker in this dataset is " + str(avg_age_of_smoker) + " years old.")
```

```python
def non_smoker_age(list):
    total_age = 0
    non_smokers = 0
    for entry in list:
        if entry[1] == 'no':
            curr_age = int(entry[0])
            total_age += curr_age
            non_smokers += 1
    avg_age_of_non_smoker = round(total_age / non_smokers, 2)
    return print("The average of a non smoker in this dataset is " + str(avg_age_of_non_smoker) + " years old.")
```

```python
# Avg Cost calculation
```

```python
def average_cost(data, total_people):
    total_cost = 0
    for key, stats in data.items():
        total_cost += float(stats['Cost'])  # Accessing the 'Cost' key directly
    avg_cost = round(total_cost / total_people, 2)
    return print("The average cost of insurance across the entire dataset is $" + str(avg_cost) + ".")
```

```python
# Determining the avg BMI of smokers vs nonsmokers
```

```python
def non_smoker_bmi(data):
    total_bmi = 0
    non_smokers = 0
    for entry in data:
        if entry[1] == 'no':
            curr_bmi = float(entry[0])
            total_bmi += curr_bmi
            non_smokers += 1
    avg_bmi_non_smoker = round(total_bmi / non_smokers, 2)
    return print("The average BMI of a non-smoker is " + str(avg_bmi_non_smoker) + ".")
```

```python
def smoker_bmi(data):
    total_bmi = 0
    smokers = 0
    for entry in data:
        if entry[1] == 'yes':
            curr_bmi = float(entry[0])
            total_bmi += curr_bmi
            smokers += 1
    avg_bmi_smoker = round(total_bmi / smokers, 2)
    return print("The average BMI of a smoker is " + str(avg_bmi_smoker) + ".")
```

```python
smoker_percentage(smokers)
smoker_age(smokers_age)
non_smoker_age(smokers_age)
average_cost(insurance_dict, total_people)
non_smoker_bmi(smokers_bmi)
smoker_bmi(smokers_bmi)
```
